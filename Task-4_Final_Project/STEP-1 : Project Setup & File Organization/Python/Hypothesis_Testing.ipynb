```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")
```

```python
df.head()
```

```python
df.columns
```

```python
male_sales = df[df["Gender"] == "Male"]["Total_Sales"]
female_sales = df[df["Gender"] == "Female"]["Total_Sales"]
```

```python
t_stat, p_value = ttest_ind(male_sales, female_sales)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the Null Hypothesis")
    print("There is a significant difference in average sales between male and female customers.")
else:
    print("Fail to Reject the Null Hypothesis")
    print("There is no significant difference in average sales between male and female customers.")
```
